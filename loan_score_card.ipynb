# Data Preprocessing

#### a. Missing value imputation and typecasting

* The analysis revealed that several columns contain missing values and unwanted characters concatenated with correct values, complicating the missing value imputation process.

* To address this, I developed a user-defined function that removes unwanted characters and performs grouped Missing Value Imputation (MVI).

* Grouped MVI involves using the mode for imputation, grouping data by customer ID so that information from other months for the same customer can be utilized to fill in missing values.

```python
def data_processing(data, mvi_groupby=None, mvi_customval=None, column=None, 
                    unwanted_value_replace=None, unwanted_value_strip=None, datatype=None):
    # Stripping unwanted values that might be at the beginning or end of the value
    if unwanted_value_strip is not None:
        if data[column].dtype == object:
            data[column] = data[column].str.strip(unwanted_value_strip)
            print(f"\nRemoved trailing and leading '{unwanted_value_strip}'")

    # Replacing unwanted value with NaN
    if unwanted_value_replace is not None:
        data[column] = data[column].replace(unwanted_value_replace, np.nan)
        print(f"\nReplaced unwanted value '{unwanted_value_replace}' with NaN")

    # Performing missing value imputation (MVI) using mode after grouping data by the specified column
    if mvi_groupby and column:
        data[column] = data[column].replace('', np.nan)
        group_mode = data.groupby(mvi_groupby)[column].transform(lambda x: x.mode().iat[0] if not x.mode().empty else np.nan)
        data[column].fillna(group_mode, inplace=True)
        print("\nMissing values imputed with group mode")

    # Performing missing value imputation using a user-provided custom value
    if mvi_customval is not None:
        data[column].fillna(mvi_customval, inplace=True)
        print(f"\nReplaced missing values with '{mvi_customval}'")

    # Changing the data type of the column based on the user-provided data type
    if datatype is not None:
        data[column] = data[column].astype(datatype)
        print(f"\nDatatype of '{column}' changed to '{datatype}'")

    print('----------------------------------------------------')
```

```python
# We initiate Missing Value Imputation (MVI) on the categorical columns

print("Processing Column: Name")
data_processing(data=bank_train, 
                column='Name',
                mvi_groupby='Customer_ID')

print("\nProcessing Column: Type_of_Loan")
data_processing(data=bank_train, 
                column='Type_of_Loan', 
                mvi_customval='Not Specified')
```

```python
#Defining a function to extract the year and month, then combine them to calculate the total number of months
def year_credit_history_to_Months(yearMonth):
    if pd.notnull(yearMonth):
        years = int(yearMonth.split(' ')[0])
        month = int(yearMonth.split(' ')[3])
        return (years * 12) + month
    else:
        return yearMonth
        
# Applying the function to the column
bank_train['Credit_History_Age'] = bank_train['Credit_History_Age'].apply(lambda x: year_credit_history_to_Months(x)).astype(float)

print('\nProcessing Column: Credit_History_Age')
data_processing(data = bank_train,
               column = 'Credit_History_Age',
               mvi_groupby = 'Customer_ID')
```

#### b. Outlier capping

```python
def outlier_removal(data, threshold=1.5):
    # Create a copy of the dataset
    outlier_data = data.copy()

    # List to store the indices of outliers
    outlier_list = []

    # Iterate through each column to identify and cap outliers in numerical columns
    for column in outlier_data.select_dtypes(include=np.number).columns:
        # Calculate Q1 and Q3 quartiles
        Q1 = outlier_data[column].quantile(0.25)
        Q3 = outlier_data[column].quantile(0.75)

        # Calculate the interquartile range (IQR)
        IQR = Q3 - Q1

        # Determine the lower and upper bounds for outliers
        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        # Identify the outliers based on the calculated bounds
        outliers = outlier_data[(outlier_data[column] < lower_bound) | (outlier_data[column] > upper_bound)]
        outlier_list.extend(outliers.index)

        # Cap the outliers to the lower and upper bounds
        outlier_data[column] = np.where(outlier_data[column] < lower_bound, lower_bound, outlier_data[column])
        outlier_data[column] = np.where(outlier_data[column] > upper_bound, upper_bound, outlier_data[column])

    # Remove duplicate indices from the outlier_list
    outlier_list = list(set(outlier_list))

    # Return the DataFrame with capped outliers and the list of outlier indices
    return outlier_data, outlier_list

# Apply the outlier removal function to the DataFrame
bank_train_clean, outliers = outlier_removal(bank_train)
```

To verify the effectiveness of outlier capping, we will generate boxplots for each column before and after capping the outliers to conduct a comparative analysis.

```python
def boxplot_outlier_comparison(data, clean_data, outlier_list):
    for column in data.select_dtypes(include = np.number).columns:
        fig, axes = plt.subplots(1,2, figsize=(10,3))
        
        # Plotting the numerical columns from the original dataframe
        sns.boxplot(data = data[column], ax = axes[0])
        
        # Plotting the numerical columns after outlier capping
        sns.boxplot(data = clean_data[column], ax = axes[1])

        axes[0].set_title(f'Before Outlier Capping: {column}')
        axes[1].set_title(f'After Outlier Capping: {column}')
        
        # Showing the plots
        plt.tight_layout()
        plt.show()

# Executing the function to compare numerical columns before and after outlier capping
boxplot_outlier_comparison(data = bank_train, clean_data = bank_train_clean, outlier_list = outliers)
```

The effectiveness of outlier capping is clear as numerous outliers, which previously skewed the data, have now been appropriately handled

```python
def crosstab_stack(data, column1, column2):
    plt.figure(figsize=(10, 6))
    #Generating a cross-tabulation for two categorical variables
    bank_crosstab = pd.crosstab(data[column1], data[column2])
    
    #Plotting the stacked bar chart
    bank_crosstab.plot(kind = 'bar', stacked = True)
    
    plt.title(f'{column1} and {column2} distribution')
    plt.xlabel(f'{column1}')
    plt.ylabel('Number of distribution')
    plt.xticks(rotation = 90)
    plt.show()
```
